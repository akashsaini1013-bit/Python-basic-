6.Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def demonstrate_mutability(x, y, lst):
    """
    Demonstrates how mutable and immutable arguments behave in Python functions.

    Args:
        x (int): Immutable argument (integer)
        y (str): Immutable argument (string)
        lst (list): Mutable argument (list)
    """
    print(f"Original values: x={x}, y='{y}', lst={lst}")

    # Immutable arguments: changes do not affect originals
    x = 200          # Reassignment creates new local int
    y = y + " modified"  # String concatenation creates new string

    # Mutable argument: changes affect original
    lst[0] = "MUTATED"
    lst.append("added")

    print(f"Inside function after changes: x={x}, y='{y}', lst={lst}")

# Example usage
original_x = 100
original_y = "Hello"
original_lst = ["unchanged"]

print("Before function call:")
print(f"  x={original_x}, y='{original_y}', lst={original_lst}\n")

demonstrate_mutability(original_x, original_y, original_lst)

print("\nAfter function call:")
print(f"  x={original_x}, y='{original_y}', lst={original_lst}")
```

11.Write a program that checks if a number is positive, negative, or zero.

```python
def number_sign_checker():
    """
    Checks if a user-input number is positive, negative, or zero.
    Handles both integer and float inputs with error handling.
    """
    try: # Added try block
        # Get user input
        num = float(input("Enter a number: "))

        print(f"\nNumber analysis for {num}:")
        print("=" * 30)

        # Check sign using comparison operators
        if num > 0:
            print("➕ Result: POSITIVE")
        elif num < 0:
            print("➖ Result: NEGATIVE")
        else:
            print("⭕ Result: ZERO")

        print(f"  Value: {num}")
        print(f"  Type: {type(num).__name__}")

    except ValueError:
        print("❌ Error: Please enter a valid number!")

# Interactive loop for multiple checks
def main():
    while True:
        number_sign_checker()
        print()

        retry = input("Check another number? (y/n): ").strip().lower()
        if retry not in ('y', 'yes'):
            print("Goodbye!")
            break

# Run the program
if __name__ == "__main__":
    main()
```

14.Write a program to reverse a string using a while loop.

```python
def reverse_string_while_loop():
    # Get user input
    original = input("Enter a string to reverse: ").strip()

    if not original:
        print("Empty string reversed is empty!")
        return

    print(f"\nOriginal: '{original}' (length: {len(original)})")
    print("Reversing with while loop:")
    print("-" * 30)

    # Reverse using while loop
    reversed_str = ""
    index = len(original) - 1  # Start from last character

    while index >= 0:
        reversed_str += original[index]
        print(f"  index {index}: '{original[index]}' → '{reversed_str}'")
        index -= 1

    print("-" * 30)
    print(f"Reversed: '{reversed_str}' ✓")

# Run the program
if __name__ == "__main__":
    reverse_string_while_loop()
```
